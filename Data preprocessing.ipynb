Here i developing a content based recommendation system,


###### content based recommendation system

A Recommender System, that provides recommendations to the user based on the item similarity format is known as a Content-Based Recommender System. So in broader terms, this type of recommender system recommends products that are similar to the products that are already liked or viewed by the users.

So here we do not need all the features in the data set ,we have to select features which gives more content about the moves

```python
#final data for preprocessing,
#creating a  function to select features which are more important for developing our model

def finaldata(feature):
    for col in data.columns:
        if not col in feature:
            data.drop(col,axis=1,inplace=True)
            
```

```python
import ast
```

```python
#fetching information 

def fetch_info(text):
    L = []
    for i in ast.literal_eval(text):
        L.append(i['name']) 
    return L 
```

```python
data['genres'] = data['genres'].apply(fetch_info)

data['keywords'] = data['keywords'].apply(fetch_info)

data['cast'] = data['cast'].apply(fetch_info)

data.head()
```

```python
#fetching first 5 actors name from cast feature

data['cast']=data['cast'].apply(lambda actor : actor[0:5])
```

```python
data.cast[0]
```

```python
data.crew[0]
```

```python
#fetching director name of each movies from crew 

def fetch_director(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L 
```

```python
data['crew'] = data['crew'].apply(fetch_director)
```

```python
data['crew']
```

```python
data.head()
```

```python
data.crew[0]
```

```python
# remove spaces because if we got one movie director is  James Cameron and the other movie by James James Cromwell,in this case 
#our recommendation will go wrong 
def space_rem(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
```

```python
data['cast'] = data['cast'].apply(space_rem)
data['crew'] = data['crew'].apply(space_rem)
data['genres'] = data['genres'].apply(space_rem)
data['keywords'] = data['keywords'].apply(space_rem)
```

```python
data.crew[0]
```

```python
data.overview[0]
```

```python
data['overview'] = data['overview'].apply(lambda x:x.split())
```

```python
data.overview[0]
```

```python
# aggrigating all the information into a single columns for getting content
data['content'] = data['overview'] + data['genres'] + data['keywords'] + data['cast'] + data['crew']
```

```python
data.head()
```

```python
#Remove all the unwanted columns
final_data = data.drop(columns=['overview','genres','keywords','cast','crew'])
```

```python
final_data
```

```python
final_data['content'] = final_data['content'].apply(lambda x: " ".join(x))
```

```python
final_data.sample(3)
```

```python
# here i used NLP for vectorizing words
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vector = cv.fit_transform(final_data['content']).toarray()
```

```python
vector
```

```python
vector.shape
```

```python
# cosine similarity is used to find the similarity between each movie
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
similarity
```

```python
final_data[final_data['original_title'] == 'The Lego Movie'].index[0]
```

```python
def recommend(movie):
    index = final_data[final_data['original_title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(final_data.iloc[i[0]].original_title)
```

```python
recommend('Avatar')
```
